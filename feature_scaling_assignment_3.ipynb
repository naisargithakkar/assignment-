    Principal component analysis (PCA) is the most widely used algorithm when dealing with dimensionality reduction.

In the example where we reduced the dimension of our training set from a two-dimension problem to a one-dimension problem, our aim was to find a line that minimizes the projection error. 
hat is, we wanted to find a line where the average orthogonal distance between the training points and the line was minimal. More formally, we wanted to find a two-dimensional unit vector u^(1) (or -u^(1)) onto which we can project the training data so as to minimize the projection error
But obviously, our data isn’t always going to be two-dimensional. So, a more generalized way of stating our problem is: Find k vectors {u^(1),u^(2),...,u^(k)} onto which to project the data, so as to minimize the projection error.

Principal Component Analysis (PCA) is an unsupervised learning algorithm technique used to examine the interrelations among a set of variables. It is also known as a general factor analysis where regression determines a line of best fit. 
The main goal of Principal Component Analysis (PCA) is to reduce the dimensionality of a dataset while preserving the most important patterns or relationships between the variables without any prior knowledge of the target variables. 

```python
#example:
import pandas as pd
import numpy as np

# Here we are using inbuilt dataset of scikit learn
from sklearn.datasets import load_breast_cancer

# instantiating
cancer = load_breast_cancer(as_frame=True)
# creating dataframe
df = cancer.frame

# checking shape
print('Original Dataframe shape :',df.shape)

# Input features
X = df[cancer['feature_names']]
print('Inputs Dataframe shape :', X.shape)
X_mean = X.mean()
 
# Standard deviation
X_std = X.std()
 
# Standardization
Z = (X - X_mean) / X_std
c = Z.cov()
 
# Plot the covariance matrix
import matplotlib.pyplot as plt
import seaborn as sns
sns.heatmap(c)
plt.show()
```

    Min-Max scaling is a technique that transforms the features of the data to a common range, usually [0, 1] or [-1, 1]. This helps to normalize the data and reduce the effect of outliers or different scales of measurement. To apply Min-Max scaling, you need to follow these steps:

For each feature, calculate the minimum and maximum values in the dataset.
For each value of the feature, subtract the minimum value and divide by the range (maximum - minimum).
Store the scaled values in a new dataset or replace the original values with the scaled ones.
For example, suppose you have a dataset with two features: price and rating. The price ranges from 100 to 500, and the rating ranges from 1 to 5. To apply Min-Max scaling to the price feature, you would do the following:
    Calculate the minimum and maximum values of the price feature: min = 100, max = 500.
For each value of the price feature, subtract the minimum value and divide by the range: (price - 100) / (500 - 100).
Store the scaled values in a new dataset or replace the original values with the scaled ones.
The scaled values of the price feature would range from 0 to 1. Similarly, you can apply Min-Max scaling to the rating feature, which would also range from 0 to 1 after scaling.
